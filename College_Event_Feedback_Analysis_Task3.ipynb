```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("student_feedback.csv")
df.head()
```

```python
if 'Unnamed: 0' in df.columns:
    df.drop(columns=['Unnamed: 0'], inplace=True)
df.isnull().sum()
```

```python
df['Overall_Score'] = df.iloc[:, 1:].mean(axis=1)
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Overall_Score'], bins=10, kde=True)
plt.title("Overall Student Satisfaction")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
avg_scores = df.iloc[:, 1:-1].mean().sort_values(ascending=False)
avg_scores
```

```python
plt.figure(figsize=(10,5))
avg_scores.plot(kind='bar')
plt.title("Average Rating per Feedback Parameter")
plt.ylabel("Average Score")
plt.xticks(rotation=90)
plt.show()
```
